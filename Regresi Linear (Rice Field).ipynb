### Regresi Linear

```python
import numpy as np #import Library
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn.linear_model import LinearRegression 
```

```python
df = pd.read_csv ('R02_rice_field.csv')  #y indepandent x dependent
df.columns = ['X','Y']
df #menampung data 
```

```python
df.corr() #mengkorelasikan data
```

```python
x_train = df['X'].values[:,np.newaxis]
y_train = df['Y'].values #menampilkan value dari masing masing variabel
```

```python
print(x_train)
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train) #linear Regression
```

```python
print('Coefficient : ' + str(lm.coef_)) #mencetak koefisien regresi 
print('Intercept : ' + str(lm.intercept_)) #mencetak 

#mencetak koefisien kemiringan dari garis regresi
#menggambarkan hubungan ;inear antar variabel 'x' dan 'y' dalam model regresi
```

```python
#prepare plot
pb = lm.predict(x_train)
dfc = pd.DataFrame({'X': df['X'],'Y':pb})
plt.scatter(df['X'],df['Y'])
plt.plot(dfc['X'],dfc['Y'],color='green',linewidth=1)
plt.xlabel('Production')
plt.ylabel('Harvest Area')
plt.show() #menampilkan statistiknya
```
